# Function to extract MDA section from the text

```python
def mdafunc(finaltext , doc_length):
    
    first_sec_un = finaltext.find('management\'s discussion and') + 1
    
    if (first_sec_un != 0):
        
        itext = finaltext[(first_sec_un)*3 : doc_length]
        first_sec_deux = itext.find('management\'s discussion and')
        
        check_text = finaltext[(first_sec_un)*3 + first_sec_deux : doc_length]
        check = check_text.find('item')
        
        try:
            mda_init = check_text[check + 5]
            mda_final = check_text[:check]
            
            return mda_final
        
        except IndexError:
            
            return 0
        
    else:
        
        return (0)
```

# Function to extract QQMDR section from the text

```python
def qqdmrfunc(finaltext , doc_length):
    
    second_sec_un = finaltext.find('quantitative and qualitative disclosures') + 1
    
    if (second_sec_un != 0):
        
        itext = finaltext[(second_sec_un)*3 : doc_length]
        second_sec_deux = itext.find('quantitative and qualitative disclosures')
        
        check_text = finaltext[(second_sec_un)*3 + second_sec_deux : doc_length]
        check = check_text.find('item')
        
        try:
            qqdmr_init = check_text[check + 5]
            qqdmr_final = check_text[:check]
            
            return qqdmr_final
        
        except IndexError:
            
            return 0
        
    else:
        
        return (0)
```

# Function to extract RF section from the text

```python
def rffunc(finaltext , doc_length):
    
    third_sec_un = finaltext.find('risk factors') + 1
    
    if (third_sec_un != 0):
        
        itext = finaltext[(third_sec_un)*3 : doc_length]
        third_sec_deux = itext.find('risk factors')
        
        check_text = finaltext[(third_sec_un)*3 + third_sec_deux : doc_length]
        check = check_text.find('item')
        
        try:
            rf_init = check_text[check + 5]
            rf_final = check_text[:check]
            
            return rf_final
        
        except IndexError:
            
            return 0
        
    else:
        
        return (0)
```

# Clean the extracted text

```python
def cleanedText(cleantext):
    
    cleantext = cleantext.lower()
    cleantext = BeautifulSoup(cleantext , features = 'lxml').get_text()
    
    tokenizer = RegexpTokenizer(r'\w+')
    cleanedtext = tokenizer.tokenize(cleantext)
    
    cleanedtextfinal = [word for word in cleanedtext if word not in stopwords.words('english')]
    
    return cleanedtextfinal    
```

# Count positive words in the cleaned text

```python
def countpositive(cleanedtext , positivewordlist):
    
    pos_words = [word for word in cleanedtext if word in positivewordlist]
    return len(pos_words)
```

# Count negative words in the cleaned text

```python
def countnegative(cleanedtext , negativewordlist):
    
    neg_words = [word for word in cleanedtext if word in negativewordlist]
    return len(neg_words)
```

# Count uncertain words in the cleaned text

```python
def countuncertain(cleanedtext , uncertainwordlist):
    
    unc_words = [word for word in cleanedtext if word in uncertainwordlist]
    return len(unc_words)
```

# Count constraining words in the cleaned text

```python
def countconstrain(cleanedtext , constrainwordlist):
    
    con_words = [word for word in cleanedtext if word in constrainwordlist]
    return len(con_words)
```

# Count complex words in the cleaned text

```python
def complexwords(cleanedtext, complexwordlist):
    
    complex_words = [word for word in cleanedtext if word in complexwordlist]
    return len(complex_words)
```

# Calculate polarity score

```python
def polarityscore(cleanedtext , posve, negve):
    
    posve = positivewordsfile()
    negve = negativewordsfile()
    
    polarity_score = (countpositive(cleanedtext,posve) - countnegative(cleanedtext,negve)) / ((countpositive(cleanedtext,posve) + countnegative(cleanedtext,negve)) + 0.000001)
    
    return polarity_score
```
